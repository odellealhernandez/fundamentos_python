## Fase 0 — Un solo diccionario (30 min)

### ¿Qué es un diccionario?

Un **diccionario** (`dict`) guarda pares **clave → valor**.  
A diferencia de una lista (acceso por posición), en un diccionario
accedemos por **nombre de campo**.

```python
# Lista: frágil — ¿qué es el índice 1?
paciente_lista = ["Germán Cerdas Valle", 28, "M", "San José"]
print(paciente_lista[1])   # 28, pero hay que saber que 1 = edad

# Diccionario: expresivo — el nombre es la documentación
paciente_dict = {
    "nombre":    "Germán Cerdas Valle",
    "edad":      28,
    "genero":    "M",
    "provincia": "San José"
}
print(paciente_dict["edad"])   # 28, y es obvio por qué
```

### Tres métodos fundamentales

```python
d = {"nombre": "Germán", "edad": 28}

d.keys()    # → dict_keys(["nombre", "edad"])
d.values()  # → dict_values(["Germán", 28])
d.items()   # → dict_items([("nombre", "Germán"), ("edad", 28)])
```

> **Conexión con S07:** `.items()` retorna pares `(clave, valor)` —
> exactamente como las tuplas que desempaquetamos con `zip()`.

> 🔮 **Predice antes de ejecutar**
>
> Si tienes `german = {"nombre": "Germán", "edad": 28}`,  
> ¿qué imprime `german["edad"] + 1`? ¿Y `german["nombre"].upper()`?
>
> _Tu predicción:_ ___

````python
# Demostración: lista vs. diccionario con el mismo paciente
paciente_lista = ["Germán Cerdas Valle", 28, "M", "San José"]
paciente_dict  = {
    "nombre":    "Germán Cerdas Valle",
    "edad":      28,
    "genero":    "M",
    "provincia": "San José"
}

print("── Con lista ──────────────────────────────")
print(f"  {paciente_lista[0]}, {paciente_lista[1]} años")  # ¿qué son 0 y 1?

print("\n── Con diccionario ────────────────────────")
print(f"  {paciente_dict['nombre']}, {paciente_dict['edad']} años")  # autoexplicado

print("\n── Tres métodos ───────────────────────────")
print(".keys()   →", list(paciente_dict.keys()))
print(".values() →", list(paciente_dict.values()))
print(".items()  →")
for clave, valor in paciente_dict.items():
    print(f"  {clave:<12}: {valor}")
````

Ya conoces las herramientas. Ahora decides cómo usarlas.

> **Consigna:** Usando los datos de la clínica, genera un reporte que responda
> **al menos 2 preguntas** sobre los pacientes. **Tú decides las preguntas.**

No hay instrucciones paso a paso. No hay salida esperada.  
Solo hay datos y herramientas.

Algunas ideas _(no son obligatorias)_:
- ¿Cuál es la edad promedio de los pacientes activos?
- ¿Cuántos pacientes hay por provincia?
- ¿Quién tiene más visitas? ¿Y menos?
- ¿Qué porcentaje del grupo está activo?

````python
# Tu reporte clínico — responde al menos 2 preguntas
print("\n--------- REPORTE CLÍNICO -------------------")
print("-"* 45)
print("Enfermedades y número de pacientes afectados")
print("-"* 45)
enfermedades = []
concurrencias = []
for paciente in pacientes:
    if paciente["enfermedad"] in enfermedades:
        concurrencias[enfermedades.index(paciente["enfermedad"])] += 1
    else:
        enfermedades.append(paciente["enfermedad"])
        concurrencias.append(1)

for enfermedad, conteo in zip(enfermedades, concurrencias):
    print(f"  {enfermedad:<20}: {conteo} pacientes")

print("-"* 45)
print("Total de pacientes afectados.")
print(f"  {sum(concurrencias)} pacientes")

# Edad promedio de los pacientes activos
print("-"* 45)
print("Edad promedio de los pacientes activos")

# Inicializar una lista para almacenar las edades de los pacientes activos
edades_activos = []

# Recorrer todos los pacientes
for paciente in pacientes:
    if paciente["activo"]:
        edades_activos.append(paciente["edad"])

suma_edades = sum(edades_activos)
numero_activos = len(edades_activos)
edad_promedio = suma_edades / numero_activos

print(f"  {edad_promedio:.2f} años")
````
