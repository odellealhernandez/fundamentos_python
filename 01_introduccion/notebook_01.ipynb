### üßº M√©todos Esenciales de Limpieza y An√°lisis
Las cadenas son objetos y tienen "poderes" (m√©todos) integrados. Para an√°lisis de datos, estos son obligatorios:

* `.strip()`: Elimina espacios vac√≠os.
* `.lower()` / `.upper()`: Normaliza may√∫sculas/min√∫sculas.
* `.split()`: Divide un texto en una lista (Crucial para CSVs).
* `in`: Verifica si un texto existe dentro de otro.

```python
texto_sucio = "   hOla MunDo  "

# 1. Limpieza y Normalizaci√≥n
limpio = texto_sucio.strip().lower()
print(f"Limpiado: '{limpio}'")

# 2. Reemplazo
fecha = "12/02/2026"
print(f"Fecha Sistema: {fecha.replace('/', '-')}")

# 3. Split (Texto a Lista) y Join (Lista a Texto)
datos_csv = "Juan,Perez,25,Costa Rica"
lista_datos = datos_csv.split(",")
print(f"Lista de datos: {lista_datos}")

# 4. Operador 'in' (B√∫squeda inteligente)
if "Costa Rica" in datos_csv:
    print("‚úÖ El usuario es de Costa Rica")
```

### üõ°Ô∏è Manejo de Errores (`try-except`)
¬øQu√© pasa si pedimos un n√∫mero y el usuario escribe "hola"? El programa explota (crash). 
Usamos `try-except` como un **casco de seguridad**.

```python
print("--- Calculadora de Edad ---")

try:
    # Intentamos ejecutar este c√≥digo peligroso
    edad_str = input("Ingresa tu edad: ")
    edad = int(edad_str)

    if edad < 0:
        print("‚ö†Ô∏è La edad no puede ser negativa.")
    elif edad < 18:
        print("Eres menor de edad üë∂")
    else:
        print("Eres mayor de edad üíº")
        
except ValueError:
    # Si falla la conversi√≥n a int, caemos aqu√≠
    print("‚ùå ¬°Error! Debes ingresar un N√öMERO entero, no letras.")
```
