## 1. Tipos de Datos y Variables en el Mundo Real üì¶

En lugar de usar variables abstractas como `x` o `y`, los profesionales de Python nombramos nuestras variables seg√∫n el contexto del negocio (PEP 8 exige nombres descriptivos en min√∫sculas separados por guiones bajos: `snake_case`).

| Tipo de Dato | Prop√≥sito | Ejemplos en un E-commerce |
|:-------------|:----------|:--------------------------|
| **`str`** (String) | Textos, nombres, c√≥digos | `'Laptop Gamer'`, `'SKU-1029'` |
| **`float`** (Decimal) | Dinero, porcentajes, pesos | `1250.50`, `0.13` (para 13%) |
| **`int`** (Entero) | Cantidades, edades, contadores | `1`, `5`, `100` |
| **`bool`** (Booleano) | Interruptores l√≥gicos (S√≠/No) | `True`, `False` |

```python
# üí° Live Coding: Inicializando el Inventario de una Tienda

# 1. Crea las variables correspondientes con sus datos y aplica PEP 8
# Python clasifica los tipos de datos por "Inferencia"
nombre_producto = 'Monitor Dell 34\"'
precio_unitario = 350000.50
unidades_en_stock = 15
aplica_envio_gratis = True

# 2. Usa print() y type() para auditar c√≥mo Python clasific√≥ cada variable
print("--- AUDITOR√çA DE TIPOS DE DATOS ---")

# Escribe tu c√≥digo aqu√≠:
print(f"Producto es: {type(nombre_producto)}")
print(f"Precio es: {type(precio_unitario)}")
print(f"Unidades stock es: {type(unidades_en_stock)}")
print(f"Envio gratis es: {type(aplica_envio_gratis)}")
```

## 2. Entrada de Datos y Casting Seguro üõ°Ô∏è

La funci√≥n `input()` es nuestra ventana para interactuar con el usuario. 
‚ö†Ô∏è **¬°Regla de Oro!** Todo lo que entra por `input()` es un texto (`str`). Si el usuario escribe `5`, Python lo ve como la letra `'5'`, no como el n√∫mero `5`.

Para hacer matem√°ticas, debemos "traducir" (Casting) usando `int()` o `float()`. ¬°Pero cuidado! Si intentamos convertir la palabra "diez" a n√∫mero, el programa explotar√°. Por eso recordaremos el `try-except` de nuestra primera clase.

```python
# üí° Live Coding: Captura y Conversi√≥n de Cantidad (Con Seguridad)

print("Bienvenido al Sistema de Ventas")

# 1. Solicita al usuario cu√°ntas unidades desea comprar
cantidad_str = input(f"¬øCu√°ntas unidades de '{nombre_producto}', desea comprar? ")

# 2. Escribe un bloque try-except para convertir cantidad_str a entero (int)
# Aseg√∫rate de validar si la cantidad es mayor a 0.

# Escribe tu c√≥digo aqu√≠:
try:
    cantidad = int(cantidad_str)
    
    if cantidad < 1:
        print("‚ùå Error: Ingrese una cantidad v√°lida mayor 0.")
    else: 
        print(f"‚è≥ Procesando {cantidad} unidades ...")
except ValueError:
    print(f"‚ùå Error cr√≠tico: por favor ingrese un n√∫mero entero v√°lido.")
    cantidad = 0
    
```

---
## 3. Operadores Matem√°ticos: El Calculador de Pedidos üßÆ

Ya tenemos n√∫meros reales en la memoria. Ahora podemos usar la calculadora interna de Python:

* `+` (Suma), `-` (Resta), `*` (Multiplicaci√≥n), `/` (Divisi√≥n Exacta)
* `//` **(Divisi√≥n Entera):** Descarta los decimales. Ideal para saber cu√°ntas cajas llenas podemos armar.
* `%` **(M√≥dulo):** Nos da el *sobrante* de una divisi√≥n. Clave para saber si un n√∫mero es par (`numero % 2 == 0`).

```python
# üí° Live Coding: C√°lculo Completo del T√≠cket y Log√≠stica

TASA_IMPUESTO = 0.13  # 13% IVA (Constantes en may√∫scula seg√∫n PEP 8)
CAPACIDAD_CAJA = 4    # Cada caja de env√≠o soporta 4 monitores

# 1. Matem√°ticas Financieras
# Calcula el subtotal, el monto_impuesto y el total_final
subtotal = cantidad * precio_unitario
monto_impuesto = subtotal * TASA_IMPUESTO
total_final = subtotal + monto_impuesto

# 2. Matem√°ticas Log√≠sticas 
# Usa // para saber cu√°ntas cajas completas salen
# Usa % para saber cu√°ntos monitores sobran sueltos
cajas_llenas = unidades_en_stock // CAPACIDAD_CAJA
monitores_sueltos = unidades_en_stock % CAPACIDAD_CAJA

# 3. Impresi√≥n Profesional con F-Strings
print("\n" + "="*30)
print("üßæ T√çCKET DE COMPRA")
print("="*30)
# Imprime los resultados con 2 decimales y formato de miles (,)
print(f"Subtotal:      ‚Ç° {subtotal:,.2f}")
print(f"IVA ({TASA_IMPUESTO:.0%}):       ‚Ç° {monto_impuesto:,.2f}")
print("="*30) 
print(f"Total:         ‚Ç° {total_final:,.2f}")
print("="*30)
print("üì¶ LOGISTICA DE ENV√çO")
print(f"   Empaque: {cajas_llenas} cajas")
print(f"   Empaque indiv:  {monitores_sueltos} unid. ")
print("="*30)
```

## 4. Operadores de Comparaci√≥n y L√≥gicos: L√≥gica de Negocio üß†

Los **operadores de comparaci√≥n** (`>`, `<`, `==`, `!=`, etc.) generan respuestas de `True/False`.
Los **operadores l√≥gicos** (`and`, `or`, `not`) combinan esas respuestas para crear el "cerebro" de nuestra aplicaci√≥n.

| Operador L√≥gico | Regla Mnemot√©cnica |
|:----------------|:-------------------|
| `and` (Y)       | **El Exigente:** TODAS las condiciones deben ser `True`. |
| `or` (O)        | **El Flexible:** Con que UNA condici√≥n sea `True`, basta. |
| `not` (NO)      | **El Rebelde:** Invierte el valor (`not True` es `False`). |

```python
# üí° Live Coding: Motor de Promociones y Descuentos

es_cliente_vip = True

# PROMOCI√ìN 1: Descuento del 10% (Uso de AND)
# Regla: Aplica SI el total es mayor a $1000 Y lleva m√°s de 20 unidades.
aplica_descuento_10 = (total_final > 1000000) and (cantidad > 20)

# PROMOCI√ìN 2: Regalo Sorpresa (Uso de OR)
# Regla: Aplica SI es Cliente VIP O SI compr√≥ exactamente 5 unidades.
aplica_regalo = es_cliente_vip or (cantidad == 5)

print("\nüéØ AN√ÅLISIS DE PROMOCIONES:")
# Imprime las dos variables de promociones aqu√≠:
print(f"Se aplica el descuento: {aplica_descuento_10}")
print(f"Lleva regalo: {aplica_regalo}")
```

## üß© DESAF√çO: ACCESO AL LABORATORIO (Validaci√≥n cruzada)

**Objetivo:** Trabajar en parejas simulando sistemas conectados. Vamos a crear el sistema de seguridad para el laboratorio de computaci√≥n. 

### ‚úÖ FASE 1: Entradas compartidas (A y B)
**Instrucci√≥n:** Ambos estudiantes deben correr exactamente este bloque en sus computadoras y escribir las mismas respuestas cuando se las pida el sistema.

```python
# =========================
# FASE 1: Entradas (A y B)
# =========================
# ¬°NO MODIFICAR ESTA CELDA! Solo ejec√∫tala e ingresa los datos.

nombre = input("Nombre del estudiante: ").strip()
edad = int(input("Edad: "))
grado = int(input("Grado (7-12): "))

hora = int(input("Hora actual (0-23): "))
tiene_carnet = input("¬øTiene carnet? (si/no): ").strip().lower() == "si"
tiene_permiso = input("¬øTiene permiso del profe? (si/no): ").strip().lower() == "si"

computadoras_disponibles = int(input("Computadoras disponibles: "))
cupos_maximos = int(input("Cupos m√°ximos del laboratorio: "))
estudiantes_dentro = int(input("¬øCu√°ntos estudiantes hay dentro ahora?: "))

equipo = input("Equipo que usar√° (pc/laptop): ").strip().lower()
tiene_cargador = input("¬øTrae cargador? (si/no): ").strip().lower() == "si"
```

---
### üë§ Estudiante A ‚Äî Auditor de Reglas Acad√©micas (Horario + Requisitos)
**Tu misi√≥n:** Validar si el estudiante cumple normas del laboratorio.

**Reglas A**
* **Edad m√≠nima:** debe ser `>= 12`.
* **Grado v√°lido:** entre 7 y 12.
* **Horario:** se permite entrar si `hora` est√° entre 7 y 17 (inclusive), o `tiene_permiso` (permiso habilita fuera de horario).
* **Carnet obligatorio:** debe tener carnet o permiso (si no tiene carnet, solo entra con permiso).
* ‚úÖ **Variable final:** Crea `normas_ok` uniendo todo con `and`.

```python
# =========================
# FASE 2A: L√≥gica A (normas)
# =========================

# 1. Valida la edad
edad_ok = edad >= 12

# 2. Valida el grado
grado_ok = grado >= 7 and grado <= 12

# 3. Valida el horario o permiso (Usa 'or')
horario_ok = (hora >= 7 and hora <= 17) or tiene_permiso 

# 4. Valida la identificaci√≥n (Usa 'or')
identificacion_ok = tiene_carnet  or tiene_permiso

# 5. Consolida todas las validaciones en una sola variable (Usa 'and')
normas_ok = edad_ok and grado_ok and horario_ok and identificacion_ok

print("\n--- RESULTADO ESTUDIANTE A ---")
print("normas_ok =", normas_ok)
print("üì¢ P√°sale a tu compa√±ero SOLO True/False:", normas_ok)
```

---
### üë§ Estudiante B ‚Äî Auditor de Capacidad y Seguridad (Cupos + Equipo)
**Tu misi√≥n:** Validar si hay espacio y condiciones t√©cnicas.

**Reglas B**
* **Capacidad:** solo entra si `estudiantes_dentro < cupos_maximos`.
* **Recursos:** debe haber computadora disponible si el equipo es "pc":
  * si `equipo == "pc"` entonces `computadoras_disponibles >= 1`
  * si `equipo == "laptop"` no necesita compu disponible
* **Seguridad t√©cnica:** si trae laptop, debe traer cargador
  * si `equipo == "laptop"` entonces `tiene_cargador == True`
* ‚úÖ **Variable final:** Crea `capacidad_ok` uniendo todo con `and`.

```python
# =========================
# FASE 2B: L√≥gica B (capacidad)
# =========================

# 1. Valida los cupos f√≠sicos
cupos_ok = estudiantes_dentro < cupos_maximos

# 2. Valida los recursos (Puedes usar if / elif / else)
if equipo == 'pc':
    recursos_ok = computadoras_disponibles >= 1
elif equipo == 'laptop':
    recursos_ok = True
else:
    recursos_ok = False

# 3. Valida la seguridad el√©ctrica
cargador_ok = True

if equipo == 'laptop':
    cargador_ok = tiene_cargador 
    
# 4. Consolida todas las validaciones en una sola variable (Usa 'and')
capacidad_ok = cupos_ok and recursos_ok and cargador_ok

print("\n--- RESULTADO ESTUDIANTE B ---")
print("capacidad_ok =", capacidad_ok)
print("üì¢ P√°sale a tu compa√±ero SOLO True/False:", capacidad_ok)
```

---
### ü§ù FASE 3: Integraci√≥n (A y B) ‚Üí un solo True/False
Ambos se preguntan el resultado del otro y lo combinan con `and` en sus respectivas computadoras.

```python
# =========================================
# FASE 3: Integraci√≥n en la PC (Estudiante A o B)
# =========================================

# 1. P√≠dele al usuario (tu compa√±ero) su resultado en texto
otro = input("Resultado de mi compa√±ero (True/False): ").strip().lower() =='true'

# 2. Convierte esa respuesta de texto a un valor Booleano (True/False)
resultado_companero_ok = normas_ok and otro

# 3. Combina TU propio resultado (normas_ok o capacidad_ok) 
# con el resultado de tu compa√±ero usando un AND l√≥gico.
acceso_aprobado = resultado_companero_ok

print("\n===========================")
print("acceso_aprobado =", acceso_aprobado)
print("===========================")
```
