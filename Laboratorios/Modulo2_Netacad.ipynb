## 2.6.11 LABORATORIO: Operadores y expresiones
**Escenario:**

La tarea es preparar un c√≥digo simple para evaluar o encontrar el tiempo final de un periodo de tiempo dado, expres√°ndolo en horas y minutos. La hora de inicio se da como un par de horas (0..23) y minutos (0..59). El resultado debe ser mostrado en la consola.
Por ejemplo, si el evento comienza a las 12:17 y dura 59 minutos, terminar√° a las 13:16.
No te preocupes si tu c√≥digo no es perfecto - est√° bien si acepta una hora invalida - lo m√°s importante es que el c√≥digo produzca una salida correcta acorde a la entrada dada.
Prueba el c√≥digo cuidadosamente. Pista: utilizar el operador % puede ser clave para el √©xito.

```python
instrucciones = """
-------------------------------------------------------------------------------------------------------------
Este c√≥digo eval√∫a o encontra el tiempo final de un periodo de tiempo dado expres√°ndolo en horas y minutos.
La hora de inicio se da como un par de horas (0..23) y minutos (0..59).
‚ÑπÔ∏è Por ejemplo: si el evento comienza a las '12:17' y dura '59' minutos, terminar√° a las '13:16'.
-------------------------------------------------------------------------------------------------------------
"""
try:
    print(instrucciones)

    hour = int(input('Hora de inicio (horas): '))
    mins = int(input('Minuto de inicio (minutos): '))
    dura = int(input('Duraci√≥n del evento (minutos): '))
    
    print('>> REVISI√ìN DE DATOS')
    print("-" * 109)    
    # validaciones de horas, minutos y duraci√≥n v√°lidas.
    val_hour = hour >= 0 and hour <= 23
    
    if val_hour:
        print(f"Hora digitada: {hour} -> ‚úÖ")
    else:
        print(f"Hora digitada: {hour} -> ‚ùå")
    
    val_mins = mins >= 0 and mins <= 59
    if val_mins:
        print(f"Minutos digitados: {mins} -> ‚úÖ")
    else:
        print(f"Minutos digitados: {hour} -> ‚ùå")
    
    val_dura = dura > 0 and dura <= 1440
    if val_dura:
        print(f"Duraci√≥n digitada: {dura} -> ‚úÖ")
    else:
        print(f"Duraci√≥n digitada: {dura} -> ‚ùå")
    
    # Evaluaci√≥n y c√°lculos -----
    val_final = val_hour and val_mins and val_dura   
    if val_final:
        print('\n >> EVALUACI√ìN Y C√ÅLCULOS')
        print("-" * 109)
        mins = mins + dura # encuentra el n√∫mero total de minutos
        print('Encuentra el n√∫mero total de minutos: ', mins)

        hour = hour + mins // 60 # encuentra el n√∫mero de horas ocultas en los minutos y actualiza las horas
        print('Encuentra el n√∫mero de horas ocultas en los minutos y actualiza las horas: ', hour)

        mins = mins % 60 # corrige los minutos para que est√©n en un rango de (0..59)
        print('corrige los minutos para que est√©n en un rango de (0..59): ', mins)

        hour = hour % 24 # corrige las horas para que est√©n en un rango de (0..23) 
        print('corrige las horas para que est√©n en un rango de (0..23): ', hour)

        print('>> FIN DE EVALUACI√ìN Y C√ÅLCULOS')

        print("-" * 109)
        print(f'‚è∞ Hora Final: {hour:02}:{mins:02}')
    else:
        print("-" * 109)
        print("\n‚ùå Error: ¬°Se ingresaron datos inv√°lidos de hora, minutos o duraci√≥n!")
    
except ValueError:
    print("-" * 109)
    print("\n‚ùå Error: ¬°Debes ingresar solo n√∫meros, no letras!")

except Exception as e:
    print("-" * 109)
    print(f"\n‚ùå Ocurri√≥ un error inesperado: {e}")
```
