## Fase 1 ‚Äî Condicionales robustos

### 1.1 if-elif-else con rangos

Las escalas de notas usan rangos. Una vez que una condici√≥n es `True`,  
Python no eval√∫a las siguientes, por eso podemos usar `>=` sin `and`:

```python
# En lugar de:
if nota >= 90 and nota <= 100:
    categoria = "Excelente"

# Podemos escribir (m√°s limpio):
if nota >= 90:         # impl√≠citamente <= 100 porque los casos superiores ya se descartaron
    categoria = "Excelente"
elif nota >= 75:
    categoria = "Bueno"
```

### 1.2 try-except: validaci√≥n de entrada

Siempre que el usuario ingrese un valor, puede equivocarse. El patr√≥n est√°ndar del curso:

```python
try:
    nota = float(input("Ingrese la nota: "))
    # ... c√≥digo que usa nota ...
except ValueError:
    print("‚ùå Ingrese un n√∫mero v√°lido.")
```

**Ejemplo completo:**

````python
# Ejemplo: clasificar una nota con try-except
try:
    nota = float(input("Ingrese la nota del estudiante (0‚Äì100): "))

    if nota < 0 or nota > 100:
        print("‚ö†Ô∏è La nota debe estar entre 0 y 100.")
    elif nota >= 90:
        print(f"Nota {nota:.1f} ‚Üí Excelente")
    elif nota >= 75:
        print(f"Nota {nota:.1f} ‚Üí Bueno")
    elif nota >= 60:
        print(f"Nota {nota:.1f} ‚Üí Suficiente")
    else:
        print(f"Nota {nota:.1f} ‚Üí Reprobado")

except ValueError:
    print("‚ùå Ingrese un n√∫mero v√°lido.")
````

## Fase 2 ‚Äî Bucle for con acumuladores

### 2.1 Iterar sobre listas

El bucle `for` recorre cada elemento de una lista, uno por uno:

```python
notas = [85, 62, 91, 74, 58]

for nota in notas:
    print(nota)
```

### 2.2 Acumuladores

Un acumulador es una variable que crece con cada iteraci√≥n:

```python
notas = [85, 62, 91, 74, 58]
suma = 0          # ‚Üê acumulador inicializado en 0

for nota in notas:
    suma += nota  # suma = suma + nota

promedio = suma / len(notas)
print(f"Promedio: {promedio:.2f}")
```

### 2.3 enumerate() ‚Äî √≠ndice + valor

`enumerate()` nos da el √≠ndice y el valor al mismo tiempo,  
sin necesidad de una variable de conteo manual:

```python
notas = [85, 62, 91]

for i, nota in enumerate(notas, start=1):
    print(f"Estudiante {i}: {nota}")
```



### <mark>Ejercicio 2.1 ‚Äî Reporte del sal√≥n</mark>

Dada la siguiente lista de notas, usamos un bucle `for` para dos tareas:

**Parte A ‚Äî Estad√≠sticas del sal√≥n:**
- Calcular el promedio general
- Contar los estudiantes aprobados (>= 70) y reprobados (< 70)

**Parte B ‚Äî Lista numerada con `enumerate()`:**  
Mostramos cada nota numerada (comenzando en 1) con su estado:
```
Estudiante  1: 92 ‚Üí Aprobado
Estudiante  2: 78 ‚Üí Aprobado
Estudiante  3: 65 ‚Üí Reprobado
...
```

````python
# Fase 2 ‚Äî Ejercicio 2.1
notas = [92, 78, 65, 88, 55, 73, 96, 82, 47, 61, 85, 70, 93, 58, 79, 100]

# Parte A: promedio, aprobados y reprobados
#Respuesta
# acumuladora de las notas aprobadas
aprobados = 0
reprobados = 0
suma = 0

for nota in notas:
    suma += nota # suma + nota
    
    if nota >= 70:
        aprobados += 1 # aprobado = aprobado + 1
        
    else:
        reprobados += 1 # reprobado = aprobado + 1

promedio = suma / len(notas)

print("------- REPORTE DE NOTAS ----------------")
#print(f"Total estudiantes: {len(notas)}")
print(f"‚úÖ Total de estudiantes aprobados: {aprobados}")
print(f"‚ùå Total de estudiantes reprobados: {reprobados}")
print("------------------------------------------")
print(f"El promedio general es: {promedio:.2f}")

# Parte B: lista numerada con enumerate()
#Respuesta
print(f"\n----- REPORTE DETALLADO -----------------")
for i, nota in enumerate(notas, start= 1):
    # estado = "Aprobado" if nota >= 70 else "reprobado"
    if nota < 0 or nota > 100:
        estado = "Excelente"
    elif nota >= 90:
        estado = "Muy bueno"          
    elif nota >= 75:
        estado = "Bueno"    
    elif nota >= 60:
        estado = "Regular"    
    else:
        estado = "Reprobado"
            
    print(f"El estudiante {i}: {nota} - estado: {estado}")

print("--------------------------------------------")    






````

## Fase 3 ‚Äî Bucle while con men√∫

### 3.1 Patr√≥n: while True + break

El patr√≥n m√°s com√∫n para men√∫s interactivos en Python:

```python
while True:
    opcion = input("Opci√≥n: ")
    if opcion == "0":

    # procesar otras opciones...
```

### 3.2 break vs continue

| Palabra clave | Efecto |
|---------------|--------|
| `break` | **Termina** el bucle completamente |
| `continue` | **Salta** el resto del ciclo actual y pasa al siguiente |

**Ejemplo contrastado:**

````python
# Diferencia entre break y continue
numeros = [10, 20, -1, 30, 40]

print("--- Con break ---")
for n in numeros:
    if n < 0:
        print("N√∫mero negativo encontrado. Deteniendo el bucle.")
        break
    print(n)

print("\n--- Con continue ---")
for n in numeros:
    if n < 0:
        print("N√∫mero negativo encontrado. Saltando.")
        continue
    print(n)
````

### 4.1 M√©todos principales

```python
notas = [85, 62, 91]
```

### 4.2 Funciones de colecciones


### 4.3 Slicing ‚Äî vista previa (Semana 07)

El slicing permite extraer partes de una lista sin modificarla:

```python
notas = [96, 91, 85, 74, 62]  # ordenadas de mayor a menor

top3    = notas[:3]   # primeros 3 ‚Üí [96, 91, 85]
ultimos = notas[-2:]  # √∫ltimos 2  ‚Üí [74, 62]
```

**Ejemplo:**

````python
# Ejemplo completo: m√©todos de listas
notas = [85, 62, 91, 74, 58]

# Lista original y estad√≠sticas
#Tama√±o, suma, m√≠nimo y m√°ximo
print(f"La lista original: {notas}")
print(f"Tama√±o de la lista: {len(notas)} elementos")
print(f"La suma de la lista: {sum(notas)}")
print(f"El M√≠nimo de la lista notas: {min(notas)}")
print(f"El M√°ximo de la lista notas: {max(notas)}")

# Ordenar y obtener Top 3
notas.sort(reverse=True) # True = DESC y False = ASC (default)
print(f"Lista de notas ordenadada mayor a menor: {notas}")
# Top 3 notas
top3 = notas[:3]
print(f"Top 3 notas: {top3}")
````

### 4.4 Lectura de un archivo externo

En las Semanas 08‚Äì10 aprenderemos a leer JSON y CSV con bibliotecas especializadas.

```python
with open("archivo.csv") as archivo:
    next(archivo)  # saltar la primera l√≠nea (encabezado)
    for linea in archivo:
```

El archivo `notas_ejemplo.csv` tiene este formato:
```
nombre,nota
Ana P√©rez,92
Carlos Mora,78
...
```

**Ejemplo guiado ‚Äî leemos el CSV paso a paso:**

````python
# Leemos notas_ejemplo.csv l√≠nea por l√≠nea
# Paso 1: abrimos el archivo con 'with open()'
# Paso 2: saltamos el encabezado con next()
# Paso 3: separamos cada l√≠nea con .split(",")
# Paso 4: acumulamos para el reporte

suma  = 0
total = 0
with open("notas_ejemplo.csv", encoding= 'utf-8') as archivo:
    next(archivo)
    for linea in archivo:
        partes = linea.strip().split(",")
        nota   = float(partes[1])
        suma  += nota
        total += 1
        print(f"estudiantes: {partes} nota: {nota:.2f}")

print(f"Promedio del sal√≥n (desde archivo): {suma/total:.2f}")
````

## Fase 5 ‚Äî Introducci√≥n a funciones

### 5.1 ¬øQu√© es una funci√≥n?

Una funci√≥n es un bloque de c√≥digo con nombre que podemos **reutilizar**:

```python
def saludar(nombre):           # def nombre_funcion(par√°metros):
    print(f"Hola, {nombre}!") # cuerpo de la funci√≥n

saludar("Ana")                 # llamada: ejecuta el bloque
saludar("Carlos")              # reutilizaci√≥n
```

### 5.2 return ‚Äî devolver un resultado

```python
def calcular_promedio(notas):
    return sum(notas) / len(notas)

promedio = calcular_promedio([85, 92, 74])
print(f"Promedio: {promedio:.2f}")  # 83.67
```

### 5.3 ¬øPor qu√© usar funciones?

| Beneficio | Explicaci√≥n |
|-----------|-------------|
| **Reutilizaci√≥n** | Escribimos el c√≥digo una vez, lo usamos muchas veces |
| **Legibilidad** | El c√≥digo principal se vuelve m√°s f√°cil de leer |
| **Organizaci√≥n** | Separamos la l√≥gica en partes peque√±as y claras |

> En las Semanas 09‚Äì13 **todas** las aplicaciones usar√°n funciones.  
> Esta fase es nuestro puente.

**Ejemplo:**

````python
# Ejemplo: funci√≥n para clasificar una nota
def clasificar_nota(nota):
    """Retorna la categor√≠a de la nota."""
    if nota >= 90:
        return "Excelente"
    elif nota >= 75:
        return "Bueno"
    elif nota >= 60:
        return "Suficiente"
    else:
        return "Reprobado"


# Usar la funci√≥n con varias notas
notas_prueba = [92, 78, 65, 45]

for nota in notas_prueba:
    categoria = clasificar_nota(nota)
    print(f"Nota {nota} ‚Üí {categoria}")
````
