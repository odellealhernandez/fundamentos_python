## Fase 1 — Condicionales robustos

### 1.1 if-elif-else con rangos

Las escalas de notas usan rangos. Una vez que una condición es `True`,  
Python no evalúa las siguientes, por eso podemos usar `>=` sin `and`:

```python
# En lugar de:
if nota >= 90 and nota <= 100:
    categoria = "Excelente"

# Podemos escribir (más limpio):
if nota >= 90:         # implícitamente <= 100 porque los casos superiores ya se descartaron
    categoria = "Excelente"
elif nota >= 75:
    categoria = "Bueno"
```

### 1.2 try-except: validación de entrada

Siempre que el usuario ingrese un valor, puede equivocarse. El patrón estándar del curso:

```python
try:
    nota = float(input("Ingrese la nota: "))
    # ... código que usa nota ...
except ValueError:
    print("❌ Ingrese un número válido.")
```

**Ejemplo completo:**

````python
# Ejemplo: clasificar una nota con try-except
try:
    nota = float(input("Ingrese la nota del estudiante (0–100): "))

    if nota < 0 or nota > 100:
        print("⚠️ La nota debe estar entre 0 y 100.")
    elif nota >= 90:
        print(f"Nota {nota:.1f} → Excelente")
    elif nota >= 75:
        print(f"Nota {nota:.1f} → Bueno")
    elif nota >= 60:
        print(f"Nota {nota:.1f} → Suficiente")
    else:
        print(f"Nota {nota:.1f} → Reprobado")

except ValueError:
    print("❌ Ingrese un número válido.")
````

## Fase 2 — Bucle for con acumuladores

### 2.1 Iterar sobre listas

El bucle `for` recorre cada elemento de una lista, uno por uno:

```python
notas = [85, 62, 91, 74, 58]

for nota in notas:
    print(nota)
```

### 2.2 Acumuladores

Un acumulador es una variable que crece con cada iteración:

```python
notas = [85, 62, 91, 74, 58]
suma = 0          # ← acumulador inicializado en 0

for nota in notas:
    suma += nota  # suma = suma + nota

promedio = suma / len(notas)
print(f"Promedio: {promedio:.2f}")
```

### 2.3 enumerate() — índice + valor

`enumerate()` nos da el índice y el valor al mismo tiempo,  
sin necesidad de una variable de conteo manual:

```python
notas = [85, 62, 91]

for i, nota in enumerate(notas, start=1):
    print(f"Estudiante {i}: {nota}")
```



### <mark>Ejercicio 2.1 — Reporte del salón</mark>

Dada la siguiente lista de notas, usamos un bucle `for` para dos tareas:

**Parte A — Estadísticas del salón:**
- Calcular el promedio general
- Contar los estudiantes aprobados (>= 70) y reprobados (< 70)

**Parte B — Lista numerada con `enumerate()`:**  
Mostramos cada nota numerada (comenzando en 1) con su estado:
```
Estudiante  1: 92 → Aprobado
Estudiante  2: 78 → Aprobado
Estudiante  3: 65 → Reprobado
...
```

````python
# Fase 2 — Ejercicio 2.1
notas = [92, 78, 65, 88, 55, 73, 96, 82, 47, 61, 85, 70, 93, 58, 79, 100]

# Parte A: promedio, aprobados y reprobados
#Respuesta
# acumuladora de las notas aprobadas
aprobados = 0
reprobados = 0
suma = 0

for nota in notas:
    suma += nota # suma + nota
    
    if nota >= 70:
        aprobados += 1 # aprobado = aprobado + 1
        
    else:
        reprobados += 1 # reprobado = aprobado + 1

promedio = suma / len(notas)

print("------- REPORTE DE NOTAS ----------------")
#print(f"Total estudiantes: {len(notas)}")
print(f"✅ Total de estudiantes aprobados: {aprobados}")
print(f"❌ Total de estudiantes reprobados: {reprobados}")
print("------------------------------------------")
print(f"El promedio general es: {promedio:.2f}")

# Parte B: lista numerada con enumerate()
#Respuesta
print(f"\n----- REPORTE DETALLADO -----------------")
for i, nota in enumerate(notas, start= 1):
    # estado = "Aprobado" if nota >= 70 else "reprobado"
    if nota < 0 or nota > 100:
        estado = "Excelente"
    elif nota >= 90:
        estado = "Muy bueno"          
    elif nota >= 75:
        estado = "Bueno"    
    elif nota >= 60:
        estado = "Regular"    
    else:
        estado = "Reprobado"
            
    print(f"El estudiante {i}: {nota} - estado: {estado}")

print("--------------------------------------------")    






````

## Fase 3 — Bucle while con menú

### 3.1 Patrón: while True + break

El patrón más común para menús interactivos en Python:

```python
while True:
    opcion = input("Opción: ")
    if opcion == "0":
        break     # sale del bucle completamente
    # procesar otras opciones...
```

### 3.2 break vs continue

| Palabra clave | Efecto |
|---------------|--------|
| `break` | **Termina** el bucle completamente |
| `continue` | **Salta** el resto del ciclo actual y pasa al siguiente |

**Ejemplo contrastado:**

````python
# Diferencia entre break y continue
numeros = [10, 20, -1, 30, 40]

print("--- Con break ---")
for n in numeros:
    if n < 0:
        print("Número negativo encontrado. Deteniendo el bucle.")
        break
    print(n)

print("\n--- Con continue ---")
for n in numeros:
    if n < 0:
        print("Número negativo encontrado. Saltando.")
        continue
    print(n)
````

## Fase 5 — Introducción a funciones

### 5.1 ¿Qué es una función?

Una función es un bloque de código con nombre que podemos **reutilizar**:

```python
def saludar(nombre):           # def nombre_funcion(parámetros):
    print(f"Hola, {nombre}!") # cuerpo de la función

saludar("Ana")                 # llamada: ejecuta el bloque
saludar("Carlos")              # reutilización
```

### 5.2 return — devolver un resultado

```python
def calcular_promedio(notas):
    return sum(notas) / len(notas)

promedio = calcular_promedio([85, 92, 74])
print(f"Promedio: {promedio:.2f}")  # 83.67
```

### 5.3 ¿Por qué usar funciones?

| Beneficio | Explicación |
|-----------|-------------|
| **Reutilización** | Escribimos el código una vez, lo usamos muchas veces |
| **Legibilidad** | El código principal se vuelve más fácil de leer |
| **Organización** | Separamos la lógica en partes pequeñas y claras |

> En las Semanas 09–13 **todas** las aplicaciones usarán funciones.  
> Esta fase es nuestro puente.

**Ejemplo:**

````python
# Ejemplo: función para clasificar una nota
def clasificar_nota(nota):
    """Retorna la categoría de la nota."""
    if nota >= 90:
        return "Excelente"
    elif nota >= 75:
        return "Bueno"
    elif nota >= 60:
        return "Suficiente"
    else:
        return "Reprobado"


# Usar la función con varias notas
notas_prueba = [92, 78, 65, 45]

for nota in notas_prueba:
    categoria = clasificar_nota(nota)
    print(f"Nota {nota} → {categoria}")
````

### <mark>Ejercicio 5.1 — Funciones del sistema de notas</mark>

Definimos dos funciones y las aplicamos con el archivo CSV:

**Función 1 — `clasificar_nota(nota)`:**  
Recibe un número y retorna la categoría: `"Excelente"`, `"Bueno"`, `"Suficiente"` o `"Reprobado"`.

**Función 2 — `reporte_salon(lista_notas)`:**  
Recibe una lista de notas y muestra:
- Promedio del salón
- Cantidad de aprobados y reprobados
- Nota más alta y nota más baja

**Aplicación:** Leemos `notas_ejemplo.csv`, clasificamos cada estudiante con `clasificar_nota()`
y llamamos a `reporte_salon()` con todas las notas.

**Salida esperada (extracto):**
```
Ana Pérez: 92 → Excelente
Carlos Mora: 78 → Bueno
...
==============================
     REPORTE DEL SALÓN
==============================
...
```

````python
def clasificar_nota(nota):
    """Fx: parámetro (número flotante)
    Retorna la categoría según la nota.
    """
    if nota >= 90:
        return "✅ Excelente"
    elif nota >= 75:
        return "✅ Bueno"
    elif nota >= 60:
        return "⚠️ Suficiente"
    else:
        return "❌ Reprobado"
    

def reporte_salon(lista_notas):
    """ Fx: parámetro (Una lista de notas)
    Genera un reporte tipo resumen con métricas estadísticas básicas
    sobre las notas de todos los estudiantes. 
    """       
    if lista_notas:
        promedio = sum(lista_notas) / len(lista_notas)
        nota_max = max(lista_notas)
        nota_min = min(lista_notas)
        
        aprobados_lista = []
        for n in lista_notas:
            if n >= 60:
                aprobados_lista.append(n)
                
        aprobados = len(aprobados_lista)
        reprobados = len(lista_notas) - aprobados

        print("\n" + "="*50)
        print("📈 REPORTE DEL SALÓN")
        print("="*50)
        print(f"🔢 Promedio del salón:   {promedio:.2f}")
        print(f"⭐ Nota más alta:        {nota_max:.2f}")
        print(f"⛔ Nota más baja:        {nota_min:.2f}")
        print(f"✅ Total aprobados:      {aprobados}")
        print(f"❌ Total reprobados:     {reprobados}")
        print("="*50)
        print(f"👦🏻👩🏻 Total de estudiantes: {len(lista_notas)}")
        print("="*50)
    else:
        print("⚠️ No se encontraron datos en el archivo.")
        
        
def RankN(lista, keynumber,  top= True, n= 3, sepchar="|"):
    """ Fx: parámetros:
            - Lista a iterar. Por ejemplo estudiantes.
            - keynumber = número entero en el cual será la lista numérica que será usada para el top.
            - top = (True/False): y si es False, será "Bottom".
            - n = será el número entero para seleccionar top 1,2,3... o bottom 1,2,3...
            - sepchar= caracter para separar la cadena resultante
    """
    if not lista:
        return "⚠️ Error: La lista está vacía."
    
    try:
        # Ordenamos la lista basándonos en la columna clave.
        # Ejemplo: [1] donde están los valores numéricos.
        # reverse=True ordena de mayor a menor (por default sería 'top')
        lista.sort(key=lambda colkey: colkey[keynumber], reverse=True)

        # Seleccionamos la sublista según sea top o bottom
        if top:
            seleccionados = lista[:n]
        else:
            # Si es bottom, los últimos N, pero los invertimos 
            # para que el "peor" salga primero en la lista
            seleccionados = lista[-n:]
            seleccionados.reverse() 

        # Resultado final como un listado.
        resultado_final = []
        for i, estudiante in enumerate(seleccionados, start=1):
            datos_con_indice = [i] + estudiante
            
            # Convertimos cada dato del estudiante a string y unimos con "|"
            linea = sepchar.join(str(dato) for dato in datos_con_indice)
            resultado_final.append(linea)
        
        # Salida de cada registro -> con un salto de línea (\n)
        return "\n".join(resultado_final)    

    except IndexError:
        return f"❌ Error: El índice '{keynumber}' no existe en los registros."
    except TypeError:
        return "❌ Error: La columna seleccionada contiene datos que no se pueden comparar (ej. mezclar texto con números)."
    except Exception as e:
        return f"❌ Ocurrió un error inesperado: {e}"

# ================================================================== 
print("\n" + "="*50)
print("📊 DETALLE DE NOTAS DE ESTUDIANTES")
print("="*50)
# ================================================================== 

lista_notas = []
lista_estudiantes = []

with open("notas_ejemplo.csv", encoding= 'utf-8') as archivo:
    next(archivo)
    for i , linea in enumerate (archivo, start= 1):
        registro = linea.strip().split(",")
        nombre_estudiante = registro[0]
        nota   = float(registro[1])
        lista_notas.append(nota)
        categ_nota = clasificar_nota(nota) # llama la función para clasificar las notas
        print(f"{i}) {nombre_estudiante}: {nota:.2f}  🢂 {categ_nota}")
        lista_estudiantes.append(registro)

# Muestra el Top 3 de las mejoras notas
print("-" * 50)
print("🏆 TOP 3 MEJORES ESTUDIANTES:")
mejores3 = RankN(lista_estudiantes, keynumber=1, top=True, n=3)
print("-" * 50)
print(mejores3)


# Muestra el Botton 3 de las peores notas
print("-" * 50)
print("📉 BOTTOM 3 PEORES ESTUDIANTES:")
peores3 = RankN(lista_estudiantes, keynumber=1, top=False, n=3)
print("-" * 50)
print(peores3)

# llama la función que genera el reporte de salón
# Esta función tiene 'prints' integrados.
reporte_salon(lista_notas)







````

````python
def obtener_ranking(listas, orden="top", n=3):
    """
    Retorna los N estudiantes con mejores o peores notas.
    lista_estudiantes: lista de listas [[nombre, nota], ...]
    orden: "top" para los mejores, "bottom" para los peores.
    n: cantidad de registros a devolver.
    """
    
    # 1. Ordenamos la lista basándonos en la nota (índice 1 del registro)
    # reverse=True ordena de mayor a menor
    listas.sort(key=lambda x: x[1], reverse=True)
    
    if orden == "top":
        listas[:n]  # Los primeros N
        for lista in listas:
            linea = "|".join(str(dato) for dato in lista)
        return print(linea)
    elif orden == "bottom":
        for lista in listas:
            linea = "|".join(str(dato) for dato in lista)
        return print(linea)
    else:
        return "⚠️ Parámetro de orden no válido."



# --- EJEMPLO DE USO ---
# Supongamos que tu lista quedó así después de leer el CSV:
listas = [
    ["Pedro", 95.0],
    ["Ana", 45.0],
    ["Luis", 88.0],
    ["Marta", 72.0],
    ["Juan", 30.0]
]

top_3 = obtener_ranking(listas, "top", 3)
print("🏆 TOP 3 ESTUDIANTES:", top_3)

peores_2 = obtener_ranking(listas, "bottom", 2)
print("📉 PEORES NOTAS:", peores_2)



````
